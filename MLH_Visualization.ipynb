```
f1 = open('Guilds.txt', 'r')
guilds={}
for i in f1:
  line = i.split('\n')[0].split(": ")
  guilds[line[0]]=int(line[1])
```

```
f2 = open('Hackers.txt', 'r')
hackers={}
for i in f2:
  line = i.split('\n')[0].split(": ")
  print(line)
  if(int(line[1])<37):
    hackers[line[0]]=int(line[1])
```

```
import matplotlib.pyplot as plt

now = [(k, v) for k, v in sorted(guilds.items(), key=lambda item: -item[1])]
graphdata = {}
for i in now[:10]:
  graphdata[i[0]]=i[1]

graphdata

colors=['lavenderblush', 'gold', 'seagreen', 'powderblue', 'bisque', 'whitesmoke', 'maroon', 'orange', 'orangered', '#838383']
plt.title(label="Top 10 Guilds at Local Hack Day: Build (until Day 2)", 
          fontsize=25, 
          color="indigo", loc='center',pad=100 , weight='bold') 
plt.pie(graphdata.values(), labels=graphdata.keys(), explode=[0.1]*len(graphdata), colors = colors, shadow=True, radius = 2)

plt.show()

```

```
from collections import Counter
c = Counter(hackers.values())
lis1= []
lis2 = []
for i in range(1,37):
  try:
    lis1.append(i)
    lis2.append(c[i])
  except:
    continue

plt.figure(figsize=(15,8))
plt.plot(lis1, lis2)
plt.title('Scores of hackers in Local Hack Day: Build (until Day 2)', fontsize=20, color='darkmagenta', weight  = 'bold')
plt.show()
```
